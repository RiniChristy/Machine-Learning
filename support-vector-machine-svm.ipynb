# **Support Vector Machine (SVM): Linear, RBF, Sigmoid & Polynomial Kernels**

## **Importing the libraries**

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

## **Importing the dataset**

```python
df = pd.read_csv('../input/social-network-ads/Social_Network_Ads.csv')
df.head()
```

## **Data Visualization**

```python
sns.jointplot(x='Age',y='EstimatedSalary',data=df, hue = 'Purchased', kind= 'scatter');
```

```python
sns.jointplot(x='Age',y='EstimatedSalary',data=df, hue = 'Purchased', kind= 'hist');
```

```python
X = df.iloc[:, [2, 3]].values
y = df.iloc[:, -1].values
```

## **Splitting the dataset into the Train set and Test set**

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)
```

## **Feature Scaling**

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

## **Training of SVM Models**

### **Training the Linear SVM model on the Train set**

```python
from sklearn.svm import SVC
classifier_Linear = SVC(kernel = 'linear', gamma=0.5, C = 1, random_state = 0)
classifier_Linear.fit(X_train, y_train)
```

### **Training the RBF Kernel SVM model on the Train set**

```python
from sklearn.svm import SVC
classifier_rbf = SVC(kernel = 'rbf',  gamma=0.5, C=1, random_state = 0)
classifier_rbf.fit(X_train, y_train)
```

### **Training the Sigmoid SVM model on the Train set**

```python
from sklearn.svm import SVC
classifier_sigmoid = SVC(kernel = 'sigmoid', gamma=0.5, C = 1, random_state = 0)
classifier_sigmoid.fit(X_train, y_train)
```

### **Training the Polynomial SVM model on the Train set**

```python
from sklearn.svm import SVC
classifier_poly = SVC(kernel='poly', gamma=0.5, C = 1, random_state = 0)
classifier_poly.fit(X_train, y_train)
```

## **Predicting the Test set results**

```python
y_pred_Linear = classifier_Linear.predict(X_test)
y_pred_rbf = classifier_rbf.predict(X_test)
y_pred_sigmoid = classifier_sigmoid.predict(X_test)
y_pred_poly = classifier_poly.predict(X_test)
```

## **Generating the Confusion Matrix and Classification report**

### **Linear SVM Model**

```python
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.metrics import accuracy_score
print('Accuracy of Linear SVM Model is ', accuracy_score(y_test, y_pred_Linear))
print('\n', '\n','Confusion Matrix of Linear SVM Model:' '\n', confusion_matrix(y_test, y_pred_Linear))
print('\n', '\n','Classification Report for Linear SVM Model:' '\n',classification_report(y_test, y_pred_Linear))
```

### **RBF Kernel SVM Model**

```python
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.metrics import accuracy_score
print('Accuracy of Kernel SVM Model is ', accuracy_score(y_test, y_pred_rbf))
print('\n', '\n','Confusion Matrix of Kernel SVM Model:' '\n', confusion_matrix(y_test, y_pred_rbf))
print('\n', '\n','Classification Report for Kernel SVM Model:' '\n',classification_report(y_test, y_pred_rbf))
```

### **Sigmoid Kernel SVM Model**

```python
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.metrics import accuracy_score
print('Accuracy of Sigmoid SVM Model is ', accuracy_score(y_test, y_pred_sigmoid))
print('\n', '\n','Confusion Matrix of Sigmoid SVM Model:' '\n', confusion_matrix(y_test, y_pred_sigmoid))
print('\n', '\n','Classification Report for Sigmoid SVM Model:' '\n',classification_report(y_test, y_pred_sigmoid))
```

### **Polynomial Kernel SVM Model**

```python
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.metrics import accuracy_score
print('Accuracy of Polynomial SVM Model is ', accuracy_score(y_test, y_pred_poly))
print('\n', '\n','Confusion Matrix of Polynomial SVM Model:' '\n', confusion_matrix(y_test, y_pred_poly))
print('\n', '\n','Classification Report for Polynomial SVM Model:' '\n',classification_report(y_test, y_pred_poly))
```

## **Computing the accuracy with k-Fold Cross Validation**

```python
from sklearn.model_selection import cross_val_score
accuracies = cross_val_score(estimator = classifier_Linear, X = X_train, y = y_train, cv = 10)
print("Accuracy of Linear SVM: {:.2f} %".format(accuracies.mean()*100))
print("Standard Deviation: {:.2f} %".format(accuracies.std()*100))
accuracies = cross_val_score(estimator = classifier_rbf, X = X_train, y = y_train, cv = 10)
print("Accuracy of RBF SVM: {:.2f} %".format(accuracies.mean()*100))
print("Standard Deviation: {:.2f} %".format(accuracies.std()*100))
accuracies = cross_val_score(estimator = classifier_sigmoid, X = X_train, y = y_train, cv = 10)
print("Accuracy of Sigmoid SVM: {:.2f} %".format(accuracies.mean()*100))
print("Standard Deviation: {:.2f} %".format(accuracies.std()*100))
accuracies = cross_val_score(estimator = classifier_poly, X = X_train, y = y_train, cv = 10)
print("Accuracy of Polynomial SVM: {:.2f} %".format(accuracies.mean()*100))
print("Standard Deviation: {:.2f} %".format(accuracies.std()*100))
```

## **Kernel Decision Boundary Visualization of Train sets**

```python
from sklearn.svm import SVC
clf1 = SVC(kernel = 'linear', gamma=0.5, C = 1, random_state = 0)
clf2 = SVC(kernel = 'rbf', gamma=0.5, C = 1, random_state = 0)
clf3 = SVC(kernel = 'sigmoid', gamma=0.5, C = 1, random_state = 0)
clf4 = SVC(kernel = 'poly', gamma=0.5, C = 1, random_state = 0)
from mlxtend.plotting import plot_decision_regions
import matplotlib.gridspec as gridspec

gs = gridspec.GridSpec(2,2)
 
fig = plt.figure(figsize=(14,10))
 
labels = ['Linear Kernel SVM Train set', 'RBF Kernel SVM Train set', 'Sigmoid Kernel SVM Train set', 'Polynomial Kernel SVM Train set'] 
for clf, lab, grd in zip([clf1, clf2, clf3, clf4], 
                         labels,
                         [(0,0), (0,1), (1,0), (1,1)]): 
 
    clf.fit(X_train, y_train)
    ax = plt.subplot(gs[grd[0], grd[1]])
    fig = plot_decision_regions(X=X_train, y=y_train, clf=clf, legend=2)
    plt.title(lab)
 
plt.show()
```
